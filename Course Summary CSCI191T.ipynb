#### Loading and analyzing Datasets
###### Using info(), describe() functions to observe dataset

```python
from sklearn import datasets,linear_model
data = sklearn.datasets.load_boston() ## loads Boston dataset from datasets library into a Pandas datafame

df = pd.DataFrame(data.data, columns=data.feature_names)
target  = pd.DataFrame(data.target, columns=['MEDV'])
fullDataset =df
fullDataset['target'] = target
fullDataset.info()
fullDataset.describe()
fullDataset.head(10)
```

### How to determine what Features to use

##### Analizing the Corelation of features and Visualizing 
###### Grouping Data into Classes with Pandas and SKLearn Pre-Processing Tools

```python
label = LabelEncoder()
fullDataset['NOX bin'] = pd.cut(fullDataset['NOX'].astype(float), 10)
fullDataset['NOX bin'] = label.fit_transform(fullDataset['NOX bin'])


fullDataset['AGE bin'] = pd.cut(fullDataset['AGE'].astype(float), 10)
fullDataset['AGE bin'] = label.fit_transform(fullDataset['AGE bin'])

fullDataset.head(10)
```

```python
sbn.countplot(fullDataset['NOX bin'], hue=fullDataset['AGE bin'])
```
